```python
### In order to create priority queues, we must use a max heap data structure

"""
The three functions below will give me where the position of the elements will be in the heap as we take it from the array.

For example, if I have a list that looks like [1, 2, 3, 4, 5, 6, 7], then then if i = 2, then the parent will be the one at the zeroth index (1),

the left child will be at 2(2)+1 = 6th index, and the right child will be at 2(2) + 2 = 7th index.

           1
         /   \
        2     3
      /  \   /  \ 
     4    5  6   7
     
input: index from an array

output: index that will help us pick the correct position of the left child, right child, and parent nodes in the MaxHeap I will be implementing.
     
     
"""
def leftChild(i):         
    return 2*i + 1   

def rightChild(i):       
    return 2*i + 2   

def parent(i):       
    return (i-1)//2  



class MaxHeap():
    """ 
    In this class, I will be coding all the methods that a maxheap data structure needs to function. 
    I also need a method that will return the max element from maxHeap.
    This is a blue print for any instances we will be creating. We will be creating the instances by sayin instanceExample = MaxHeap().
    
    """  
  
    def __init__(self):        
        self.listValuesofHeap = []  # initializing it with empty list so that I can store the values later.
        self.numberOfKeysInHeap = 0  # helps me to keep track of the number of keys to see if the Heap is out of elements/keys.

   
    def heapPush(self, inputKey):    #   the user should use the follwing command to push element to this heap ==> instance.heapPush(inputKey)
        
        '''
        Here, I will be pushing the elements from the list into the heap. IN this method, I will be doing the following three things:-
        
        Start by appending negative infinity as I am planning to implement a MaxHeap. (the maxElemet will be the atmost parent and for that we need to start the comparison with a very small number.)
        
        I will call the keyTracker() function. So that the keys will be swapped according to their value and their weight of taskImportance.
        
        Then, I will increament the number of keys, as one key was added to the Heap.
        '''
      
        self.listValuesofHeap.append(-float("inf"))  
        #I will start by appending negative infinity as I will be implementing a maxHeap. As I will be comparing the the value at the parent index of self.listValuesofHeap with this value(with respect to their corresposning task importance), it's a good idea to to make it negative infinity as no amound of smaller input will be less than negative infinity.(I would have made it very negative number like -111111111111), but what if someone input an even lower number? 
        self.keyTracker(self.numberOfKeysInHeap,inputKey) # calls the method below, which will do the swapping of elements according to their value and weigh of task importance.
        self.numberOfKeysInHeap+=1  # increments the value as one key was added. 
        



  
    #   the user can use the following command to access the method below ==> instance.keyTracker(indexVal, inputKey)
    def keyTracker(self, indexVal, inputKey): # this method updates the value of each key in the maxHeap with a larger value as indicated in the {while statement.}

        self.listValuesofHeap[indexVal] = inputKey # this line is very important becuase it will give me acess for the priority queue
        #without it, the while loop will be faced with an error as self.listValuesofHeap[parent(indexVal)] will just be a float. And a float object has no attribute as taskImportance.
        #so inorder to access the taskImportance and while making decision for the keys in maxHeap, I need to specify that my inputKey is also self.listValuesofHeap[indexVal].
        
        while indexVal > 0 and self.listValuesofHeap[parent(indexVal)].taskImportance < self.listValuesofHeap[indexVal].taskImportance:
            secondIndexVal = parent(indexVal)
            temporaryMemory = self.listValuesofHeap[secondIndexVal]
            self.listValuesofHeap[secondIndexVal] = self.listValuesofHeap[indexVal]
            self.listValuesofHeap[indexVal] = temporaryMemory
            indexVal = secondIndexVal    
            
 
    #   The user can use the folowing command to acess the method below ==> instance.heapify(indexOfRootNode)   
    
    # It will be taking the index of the root Node and it will heapify the tree that will result in the the max element ot be at the root of the tree. 
    
    def heapify(self, indexOfRootNode):
        left = leftChild(indexOfRootNode) # calling the function to find the index of the left child for the given indexOfRootNode
        right = rightChild(indexOfRootNode) # calling the function to find the index of the right child for the given indexOfRootNode. 
        
         
        if left <= (self.numberOfKeysInHeap-1) and self.listValuesofHeap[left].taskImportance > self.listValuesofHeap[indexOfRootNode].taskImportance:
            largest = left # if the left child is greater than the root Node, it's index will be my largest index value[I also considered the taskImportance while making this decision]
        else:
            largest = indexOfRootNode # otherwise, my largest will be the index of  root node. 
        if right <= (self.numberOfKeysInHeap-1) and self.listValuesofHeap[right].taskImportance > self.listValuesofHeap[largest].taskImportance:
            largest = right # now, I will compare the right with the largest(whatever that was), and the largest will be the overall largest value.
        if largest != indexOfRootNode: 
            # if my largest is different from the rootNode, it means the largest number was not at the root and this is wrong for maxHeap, so I wll exchane the position of the root node with that of the overall largest. 
            
            
            self.listValuesofHeap[indexOfRootNode], self.listValuesofHeap[largest] = self.listValuesofHeap[largest], self.listValuesofHeap[indexOfRootNode]
            self.heapify(largest) # this will bring the max element to the root for easier acess as we want in maxHeap.

    #the method below will takeout the largets element from the heap the same way the built-in heappop() function does in our python.
    #   the  user can use the follwing command to access the method below ==> instance.heapPop(รท)
    def heapPop(self):
        if self.numberOfKeysInHeap < 1:
            print("zero keys left in the priority queue")
        maxElementOfHeap = self.listValuesofHeap[0] # storing the maxElement before swapping it with the last element
        self.listValuesofHeap[0] = self.listValuesofHeap[-1] 
        self.listValuesofHeap.pop() # remove the key of the largest key.
        self.numberOfKeysInHeap-=1 # decrementing the number fo key as we have just popped one element out. 
        self.heapify(0) # heapifying using the index of the root Node.
        return maxElementOfHeap
    
    # The method below will give us the highest key in out maxHeap as needed. Since the maxElement will be at the root index(index 0), it will give us acess to that.
    
    #   the user can use the follwoing command to access the method below ==> self.maxElementOfHeap()
    def maxElementOfHeap(self):              
        return self.listValuesofHeap[0]     
    
```
